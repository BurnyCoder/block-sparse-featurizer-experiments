# Block-Sparse Featurizer Workbench

This launcher reuses the packaged app builder. Gradio displays `Blocks.launch()` inline by default in notebooks: https://www.gradio.app/docs/gradio/blocks#blocks-launch.

```python
from bsf_experiments.config import load_app_config
from bsf_experiments.ui import build_app

config = load_app_config()
app = build_app(config)
app.queue(default_concurrency_limit=1)
app.launch(
    server_name="127.0.0.1",
    server_port=config.port,
    share=False,
    inline=True,
    allowed_paths=[str(config.output_dir)],
    blocked_paths=[str(config.env_file)],
    max_file_size=config.max_upload_mb * 1024 * 1024,
)
```
